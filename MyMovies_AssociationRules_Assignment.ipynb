```python
# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mlxtend.frequent_patterns import apriori,association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
movie=movies.iloc[:,5:]
movie
```

## Apriori algorithm

### Association rules with 10% Support and 70% confidence

```python
# with 10% support
frequent_itemsets=apriori(movie,min_support=0.1,use_colnames=True)
frequent_itemsets
```

```python
# 70% confidence
rules=association_rules(frequent_itemsets,metric='lift',min_threshold=0.7)
rules
```

### Computing the antecedent length

```python
rules["antecedent_len"] = rules["antecedents"].apply(lambda x: len(x))
rules
```

#### Records with minimum length 1

```python
frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(lambda x: len(x))
frequent_itemsets[ (frequent_itemsets['length'] == 1) &
                   (frequent_itemsets['support'] >= 0.10)]
```

#### Antecedent length = 1

```python
rules[ (rules['antecedent_len'] ==1 ) &
       (rules['confidence'] > 0.75) &
       (rules['lift'] > 1.2) ]
```

#### Records with minimum length 2

```python
frequent_itemsets[ (frequent_itemsets['length'] == 2) &
                   (frequent_itemsets['support'] >= 0.10)]
```

#### Antecedent length = 2

```python
rules[ (rules['antecedent_len'] ==2 ) &
       (rules['confidence'] >= 0.80) &
       (rules['lift'] > 1.8)
    ]
```

#### Records with minimum length 3

```python
frequent_itemsets[ (frequent_itemsets['length'] == 3) &
                   (frequent_itemsets['support'] >= 0.10)]
```

#### Antecedent length = 3

```python
rules['antecedent_len'] ==3 # Since no antecedent len==3 no rules can be formed
```

### Forming rules for entire dataset

```python
# Lift Ratio > 1 is a good influential rule in selecting the associated transactions
rules[rules.lift>1]
```

```python
# visualization of obtained rule
plt.scatter(rules['support'],rules['confidence'])
plt.xlabel('support')
plt.ylabel('confidence')
plt.show()
```

```python
%matplotlib inline
rules[['support', 'confidence','lift' ]].plot()
```
